# Analysis of Shooting Data

```python
import pandas
import numpy as N
import tqdm
import matplotlib.pyplot as pyplot
import seaborn
from pathlib import Path
import geopy

seaborn.set_style('whitegrid')
```

```python
dfShooting = pandas.read_csv('data/NYPD_Shooting_Incident_Data__Historic_.csv')
dfShooting
```

```python

geolocator = geopy.Nominatim(user_agent='stat')
def get_zipcode(lat, long):
    location = geolocator.reverse((lat, long))
    return location.raw['address']['postcode']
zipcodes = [get_zipcode(lat, long) for lat, long in tqdm.tqdm(list(zip(dfShooting['Latitude'], dfShooting['Longitude'])))]
dfShooting['zip code'] = zipcodes
```

```python
dfShooting['year'] = dfShooting['OCCUR_DATE'].apply(lambda x:x[-4:])
```

```python
dfShooting.to_csv('processed/shooting.csv', index=False)
```
